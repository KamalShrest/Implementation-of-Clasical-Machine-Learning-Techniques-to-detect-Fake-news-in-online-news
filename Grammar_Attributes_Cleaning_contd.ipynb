```python
#STEMMING
#LEMMATIZATION

import pandas as pd
import numpy as np
import nltk
import nltk.data
from numpy import dot
from nltk.tokenize import word_tokenize, sent_tokenize
#from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
#sw = set(stopwords.words("english"))
```

```python
data=pd.read_csv('Real.csv')
data = data[['title', 'description']]
title, description = data['title'], data['description']
x = len(title)
y = len(title)
data
#data[['title', 'label']]
```

```python
#STEMMEING FOR title
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
stemmed_title=[]
for i in range(x): #Iterates from 0 to length of title
    words_title = word_tokenize(title[i]) #Tokenizing each row
    for j in words_title: #Iterate Through each list element(generated from title row)
        print(ps.stem(j), end =" ") #each row sentence or paragraph is seperated by a comma in the result below
```

```python
#STEMMING FOR description
#Similar to Above
ps = PorterStemmer()
stemmed_description=[]
for i in range(x):
    words_title = word_tokenize(description[i])
    for j in words_title:
        print(ps.stem(j), end=" ")
```

```python
#LEMMATIXATION FOR title
lemma_title = WordNetLemmatizer()
lemalized_title = [] #list created
for i in range(x): #Iterate though 0 to length of title
    words_title = word_tokenize(title[i]) #Tokenizing
    for j in words_title: 
        print(lemma_title.lemmatize(j), end=" ") #Lemmatixe and print
```

```python
#LEMMATIZATION FOR description
from nltk.stem import WordNetLemmatizer
lemma_description = WordNetLemmatizer()
lemalized_discription = []
for i in range(x):
    words_description = word_tokenize(description[i])
    for j in words_description:
        print(lemma_description.lemmatize(j), end=" ")
```
